## 8. Examine missing values (2+2+3=7pt)

### 8.1 Show a list with column wise count of missing values and display the list in count wise descending order.

```python
missing_values = df.isnull().sum().sort_values(ascending=False)
print("\033[1;4m" + 'Feature' + "\033[0m" + '\t\t\t' + "\033[1;4m" + 'Total Missing' + "\033[0m")
print(missing_values)
```

### 8.2 Show columnwise percentage of missing values.

```python
percent_missing = (df.isnull().sum().sort_values(ascending=False)  / len(df)) * 100
print("\033[1;4m" + 'Feature' + "\033[0m" + '\t\t\t' + "\033[1;4m" + 'Percentage Missing' + "\033[0m")
print(percent_missing)
```

### 8.3 Display a bar plot to visualize only the columns with missing values and their percentage count.

```python
percent_missing[percent_missing>0].plot.bar(ylabel='percentage (%)',title='Percentages of columns with missing values ',figsize=(5,5))
```

## 9. Label Encoding : Copy the dataframe to a new one. Then using scikitlearn's Label Encoder, transform the "status" column to 0-1. (5pt)

```python
from sklearn.preprocessing import LabelEncoder
encoded_df = df
encoder = LabelEncoder()
encoded_df['status'] = encoder.fit_transform(encoded_df['status'])
display(encoded_df)
```

## 10. Correlation: Use seaborn's heatmap to visualize the correlation between numeric features. (3pt)

```python
corr = encoded_df.corr()
fig, ax = plt.subplots(2, 1, figsize=(10,15))
sns.heatmap(ax=ax[0], data=corr)
sns.heatmap(ax=ax[1], data=corr, cmap='RdBu', annot=True, annot_kws={'fontsize':8, 'fontweight':'bold'})
plt.show()
```

## 11.1 Use seaborn's countplot to visualize relationship between "*state_code*" and "*labels*". Comment on which state produced majority of successful startups (3pt)

```python
ax = sns.countplot(x='state_code',hue='labels',data=encoded_df)
sns.move_legend(ax,'upper right')
sns.set(rc={'figure.figsize':(15,5)})
print('CA produced majority of successful startups')
```

## 11.2 Use seaborn's countplot to visualize relationship between "*milestones*" and "*labels*". Comment on which milestone made the statistically highest number of successful startups (3pt)

```python
ax = sns.countplot(x='milestones',hue='labels',data=encoded_df)
sns.move_legend(ax,'upper right')
sns.set(rc={'figure.autolayout':True})
print('Milestone 2 made the statistically highest number of successful startups')
```
